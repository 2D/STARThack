```python
import socket
TCP_IP = '192.168.254.3'
TCP_PORT = 1212
buffer_size = 1024

message = bytes('%R1Q,9081:0,150.0,200.0', 'utf-8')


s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.settimeout(10)
try:
    s.connect((TCP_IP, TCP_PORT))
    s.send(message)
    data = s.recv(buffer_size)
    s.close()
    print ("received data:", data)
except socket.timeout as e:
    print (e)
```

```python
#!/usr/bin/env python3

import socket

HOST = '192.168.254.3'  # Standard loopback interface address (localhost)
PORT = 1212        # Port to listen on (non-privileged ports are > 1023)

with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
    s.bind(HOST, PORT)
    s.listen()
    conn, addr = s.accept()
    with conn:
        print('Connected by', addr)
        while True:
            data = conn.recv(1024)
            if not data:
                break
            conn.sendall(data)
```

```python
import socket
import os
import time

TCP_IP = '192.168.254.3'
TCP_PORT = 1212
buffer_size = 1024

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.settimeout(5)
arr=[]
try:
    s.connect((TCP_IP, TCP_PORT))
    #set target
    s.send(b'%R1Q,9027:0.52,1.5,0,0,0\r\n')
    s.send(b'%R1Q,17021:1\r\n')
    for i in range(1,500,100):
        time.sleep(2)
        #move
        command = '%R1Q,9081:0,150.0,200.0\r\n'
        command = str.encode(command)
        s.send(command)
        time.sleep(2)
        #measure
        command = '%R1Q,17017:2\r\n'
        #command = '%R1Q,17017:1\r\n'
        command = str.encode(command)
        s.send(command)
        #record measurement
        data = s.recv(4096)
        print ('recived data = ', data.decode() )
        time.sleep(2)
        arr.append(data.decode() )
    #print(s.recv(4096))
    s.close()
    print ("success")
except socket.timeout as e:
   print(e)
```

```python
import socket
import os
import time

TCP_IP = '192.168.254.3'
TCP_PORT = 1212
buffer_size = 1024

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.settimeout(1)
try:
    s.connect((TCP_IP, TCP_PORT))
    for i in range(1,628,1):
        time.sleep(3)
        #command = '%R1Q,9027:0.0,0.0,0,0,0\r\n'
        x = str((i)/10)
        y = str(3.14/2)
        command = '%R1Q,9027:'+x+','+y+',0,0,0\r\n'
        command = str.encode(command)
        s.send(command)
        #print(s.recv(4096))
        #measure
        time.sleep(2)
        command = '%R1Q,17017:2\r\n'
        #command = '%R1Q,17017:1\r\n'
        command = str.encode(command)
        s.send(command)
        #record measurement
        data = s.recv(4096)
        print ('recived data = ', data.decode() )
        arr.append(data.decode() )
        time.sleep(3)
    s.close()
    print ("success")
except socket.timeout as e:
   print(e)
```
